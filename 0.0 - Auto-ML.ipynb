```python
from autosklearn.classification import AutoSklearnClassifier
print(autosklearn.__version__)
import pandas as pd
from sklearn.preprocessing import Normalizer
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score
from autosklearn.metrics import (accuracy,
                                 f1,
                                 roc_auc,
                                 precision,
                                 average_precision,
                                 recall,
                                 log_loss)
```

```python
df_train = pd.read_csv('/content/drive/MyDrive/Redes Neurais/Github/Redes-Neurais/Titanic/train_tratado.csv')

train = df_train.drop(columns = ['Transported'])
y = df_train['Transported']


### Normalizar

transformer = Normalizer().fit(train) 
transformer
train=transformer.transform(train)


#Splitar em treinamento e teste

x_train, x_test, y_train, y_test = train_test_split(train, y, test_size=0.3)
```

```python
#define the model
automl = AutoSklearnClassifier(time_left_for_this_task=300,per_run_time_limit=30)
#train the model
automl.fit(x_train, y_train )
#predict
y_pred=automl.predict_proba(x_test)
# score
score=roc_auc_score(y_test,y_pred[:,1])
print(score)
```

```python
print(automl.sprint_statistics())
```

```python
automl.leaderboard()
```

```python
automl.leaderboard(detailed = True, ensemble_only=False)
```

```python
from autogluon.tabular import TabularDataset, TabularPredictor
import pandas as pd
```

```python
df_train = pd.read_csv('/content/drive/MyDrive/Redes Neurais/Github/Redes-Neurais/Titanic/train_tratado.csv')
```

```python
train_data = df_train.iloc[:4600]
test_data = df_train.iloc[4600:]
y_test = test_data['Transported']
label = 'Transported'
```

```python
redictor = TabularPredictor(label=label).fit(train_data)
```

```python
y_pred = redictor.predict(test_data)
print("Predictions:  \n", y_pred)
perf = redictor.evaluate_predictions(y_true=y_test, y_pred=y_pred, auxiliary_metrics=True)
```

```python
redictor.leaderboard(test_data, silent=True)
```
